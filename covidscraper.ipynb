```python
# Imports
import requests
from bs4 import BeautifulSoup as bs
```

```python
# request the website
r = requests.get('https://ncov2019.live/data')

# get the html from the request
page = r.text

# create a beautiful soup object and use lxml to proccess the html
soup = bs(page, 'lxml')
```

```python
# get the body html
soup.find('body')
```

```python
# get world data table
table = soup.find('table', attrs={'id':'sortable_table_world'})
table
```

```python
# get table rows
tableRows = table.findAll('tr')
tableRows
```

```python
tableCols = tableRows[1].findAll('td')
tableCol = tableCols[0]
tableCol
```

```python
tableCol.text
```

```python
tableCol.text.strip()
```

```python
tableCol2 = tableRows[1].findAll('td')[2].text.strip()
tableCol2
```

```python
for row in tableRows[1:]:
    if 'United States' == row.find('td').text[2:].strip():
        columns = row.findAll('td')
        print(f'Confirmed: {columns[1].text.strip()}')
        print(f'Changes: {columns[2].text.strip()}')
        print(f'Critical: {columns[3].text.strip()}')
```
